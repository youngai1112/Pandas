# `merge`

`concat`과 `merge`는 단순하게 합쳐지는 목적 또는 특정 기준(`index`)으로 합치느냐에 따라 용도가 달라진다.
* `concat` : `row`나 `column` 기준으로 단순하게 이어 붙이기
    * 이어 붙여지는 기준 : 열 인덱스(axis=0), 행 인덱스(axis=1)
* `merge` : 특정 고유한 키(`unuque_id`)

```python
df_idol1.head()
```

```python
df_idol2.head()
```

"이름" 컬럼을 기준으로 df1, df2를 병합

**`pd.merge(left_df, right_df, on = "기준 컬럼", how="left")`**
* how에는 `left`, `right`, `inner`, `outer` 4가지 병합 방식을 지원

```python
df2_right = df_idol2.drop([1, 3, 5, 7, 9])
df2_right.shape
```

```python
df2_right = df2_right.reset_index(drop=True)
df2_right.head()
```

* `left` 옵션을 부여하면 왼쪽 데이터 프레임(`left_df`)에 키 값이 존재하면 해당 데이터를 유지
    * `right_df`에 데이터가 없으면 NaN값이 들어간다. 

```python
pd.merge(df_idol1, df2_right, on="이름", how = "left") # 1, 3, 5, 7, 9 삭제 => NaN값 처리
```

`how = right`로 설정하면?
* 오른쪽 데이터 프레임을 기준으로 합쳐진다.
* 왼쪽 데이터 프레임이 데이터의 개수가 더 많다.(15개) 오른쪽 데이터 프레임(10개)

```python
pd.merge(df_idol1, df2_right, on="이름", how="right") # right데이터 기준으로
```

## `inner`, `outer`
* `inner` : 두 데이터 프레임에 모두 키 값이 존재하는 경우에 병합
    * 겹치는 키값만 살아 남는다.
* `outer` : 하나의 데이터 프레임에 키 값이 존재하는 경우 모두 병합
    * 양 측 모두의 키값이 모두 살아남는다.
   

```python
pd.merge(df_idol1, df2_right, on="이름", how="inner")
```

```python
pd.merge(df_idol1, df2_right, on="이름", how = "outer") # 서로에게 없는 부분은 전부 NaN값
```
