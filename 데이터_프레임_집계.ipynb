```python
import pandas as pd
```

```python
df_idol = pd.read_csv("./data/korean-idol.csv")
df_idol.head()
```

# 피벗 테이블(pivot table)
* 데이터 프레임의 데이터를 인덱스로 사용하기 위한 테이블
    * 기준을 데이터로 잡고 집계
* 항목(카테고리) 데이터를 컬럼이나 행의 인덱스로 만들기 위함
* 왼쪽에 나타나는 인덱스를 **행 인덱스**, 상단에 나타나는 인덱스를 **열 인덱스(컬럼 이름)**

```python
pd.pivot_table(
    df_idol, # pivot table을 만들 데이터 프레임
    index="소속사", # 행 인덱스 설정,
    columns="성별", # 열 인덱스 설정
)
```

```python
pd.pivot_table(
    df_idol,
    index="그룹",
    columns="혈액형",
    values="키" # 집계 대상 컬럼
)
```

```python
import numpy as np

pd.pivot_table(
    df_idol,
    index="그룹",
    columns="혈액형",
    values="브랜드평판지수",
    aggfunc=[np.sum, np.mean]
)
```

# 그룹핑
* `groupby` 함수를 활용
* 데이터를 **그룹으로 묶어서 분석할 때 유용**
* **~별로 집계**

```python
df_idol.groupby("소속사") # 이 상태에서 필요한 집계를 수행
```

집계의 종류
* count() - 개수
* sum() - 합계
* mean() - 평균
* var() - 분산
* std() - 표준편차
* min() - 최소값
* max() - 최대값

```python
# nan값을 제외한 데이터의 개수를 세어준다.
df_idol.groupby("소속사").count()
```

```python
# count를 제외한 기타 집계 함수는 숫자형식의 데이터에만 적용이 가능
df_idol.groupby("그룹").mean()
```

```python
# 혈액별 키의 평균
df_idol.groupby("혈액형")["키"].mean()
```
