# apply
Series, DataFrame에 **구체적이고 복잡한 로직**에 의해 데이터를 편집하고 싶을 때 사용
* 단순 집계 뿐만이 아닌, 각 데이터에 프로그래밍적인 효과
    * 데이터에 대한 여러 조건 검사
    * 형태소 분리
    * 복잡한 조건에 의한 데이터 편집 등등..
* `apply(함수)`
    * apply 안에 들어가는 함수는 데이터 편집 후 반드시 리턴

```python
# 남자는 1, 여자는 0으로 설정

df2 = df.copy()

man_mask = df2["성별"] == "남자"
woman_mask = df2["성별"] == "여자"

df2.loc[man_mask, "성별"] = 1
df2.loc[woman_mask, "성별"] = 0

df2.head()
```

apply를 사용하면? --map 함수와 유사

```python
df["성별_NEW"] = df["성별"].apply(lambda x : 1 if x == "남자" else 0)
df

# df["성별"]의 데이터 하나하나가 
#.apply(lambda x 로 들어감
#       : 1 if x == "남자" else 0)
```

```python
df["키/2"] = df["키"].apply(lambda x : x / 2 )
df
```
